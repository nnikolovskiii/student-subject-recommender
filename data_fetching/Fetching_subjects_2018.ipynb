```python
import fitz  # PyMuPDF
import requests
import re
```

```python
def split_text_by_pattern(text, pattern):
    return re.split(pattern, text)
```

```python
def pdf_to_text_from_url(pdf_url):
    try:
        response = requests.get(pdf_url)
        pdf_document = fitz.open(stream=response.content, filetype="pdf")

        text = ""
        for page_num in range(pdf_document.page_count):
            page = pdf_document[page_num]
            text += page.get_text()

        pdf_document.close()
        return text
    except Exception as e:
        print(f"Error: {e}")
        return None
```

```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm

urls = ['https://www.finki.ukim.mk/mk/content/%D0%B8%D0%BD%D1%82%D0%B5%D1%80%D0%BD%D0%B5%D1%82-%D0%BC%D1%80%D0%B5%D0%B6%D0%B8-%D0%B8-%D0%B1%D0%B5%D0%B7%D0%B1%D0%B5%D0%B4%D0%BD%D0%BE%D1%81%D1%82'
,'https://www.finki.ukim.mk/mk/content/%D0%BA%D0%BE%D0%BC%D0%BF%D1%98%D1%83%D1%82%D0%B5%D1%80%D1%81%D0%BA%D0%B0-%D0%B5%D0%B4%D1%83%D0%BA%D0%B0%D1%86%D0%B8%D1%98%D0%B0'
,'https://www.finki.ukim.mk/mk/content/%D0%BA%D0%BE%D0%BC%D0%BF%D1%98%D1%83%D1%82%D0%B5%D1%80%D1%81%D0%BA%D0%B8-%D0%BD%D0%B0%D1%83%D0%BA%D0%B8'
,'https://www.finki.ukim.mk/mk/content/%D0%BA%D0%BE%D0%BC%D0%BF%D1%98%D1%83%D1%82%D0%B5%D1%80%D1%81%D0%BA%D0%BE-%D0%B8%D0%BD%D0%B6%D0%B5%D0%BD%D0%B5%D1%80%D1%81%D1%82%D0%B2%D0%BE'
,'https://www.finki.ukim.mk/mk/content/%D0%BF%D1%80%D0%B8%D0%BC%D0%B5%D0%BD%D0%B0-%D0%BD%D0%B0-%D0%B8%D0%BD%D1%84%D0%BE%D1%80%D0%BC%D0%B0%D1%86%D0%B8%D1%81%D0%BA%D0%B8-%D1%82%D0%B5%D1%85%D0%BD%D0%BE%D0%BB%D0%BE%D0%B3%D0%B8%D0%B8'
, 'https://www.finki.ukim.mk/mk/content/%D1%81%D0%BE%D1%84%D1%82%D0%B2%D0%B5%D1%80%D1%81%D0%BA%D0%BE-%D0%B8%D0%BD%D0%B6%D0%B5%D0%BD%D0%B5%D1%80%D1%81%D1%82%D0%B2%D0%BE-%D0%B8-%D0%B8%D0%BD%D1%84%D0%BE%D1%80%D0%BC%D0%B0%D1%86%D0%B8%D1%81%D0%BA%D0%B8-%D1%81%D0%B8%D1%81%D1%82%D0%B5%D0%BC%D0%B8']


program = [ 'IMB18','IE18', 'KN18', 'KI18','PIT18', 'SIIS18' ]

names = []
programs = []
codes = []
contents = []
semesters = []

for i, url in enumerate(tqdm(urls[0:1], desc="Scraping Progress")):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    li = soup.find_all('div', class_='view-grouping-content')
    for j,elem in enumerate(li[0:1]):
      containers = elem.find_all('div', class_='table-responsive')
      elems = containers[0].find_all('a')
      li1 = [elem["href"] for elem in elems]
      for link in li1:
        full_url = f"https://www.finki.ukim.mk{link}"
        response = requests.get(full_url)
        soup = BeautifulSoup(response.text, 'html.parser')
        name = soup.find_all("li", class_='abreadcrumb')
        articles = soup.find_all("article")
        divs = articles[0].find_all("div", class_='field-item even')
        pdf_link = divs[3].find_all("a")[0]["href"]
        text_content = pdf_to_text_from_url(pdf_link)
        split_text = split_text_by_pattern(text_content, r'\d+\.')

        split_text = [part.strip() for part in split_text if part.strip()]

        print("*"*50)
        for part in split_text:
          if "Предуслови за запишување на \nпредметот" in part:
            print(part.replace("Предуслови за запишување на \nпредметот", '', 1) )
            print("-"*30)
          if "Цели на предметната програма (компетенции):" in part:
            print(part.replace("Цели на предметната програма (компетенции):", '', 1) )
            print("-"*30)
          if "Содржина на предметната програма:" in part:
            print(part.replace("Содржина на предметната програма:", '', 1) )
            print("-"*30)

        print("*"*50)

        if len(divs)<2:
          pass
        else:
          codes.append(divs[0].text)
          contents.append(divs[2].text)
          semesters.append(j+1)
          programs.append(program[i])
          names.append(name[0].text.split(">")[0])
```

```python
import pandas as pd

df = pd.DataFrame({
    'Program': programs,
    'Name': names,
    'Code': codes,
    'Content': contents,
    'Semester': semesters
})

df.head()
```
