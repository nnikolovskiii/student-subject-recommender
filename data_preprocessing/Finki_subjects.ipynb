```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')
```

```python
sentences = ["This is an example sentence", "Each sentence is converted"]
embeddings = model.encode(sentences)
embeddings.shape
```

```python
import tqdm
embeddings = []
names = []

for index, row in tqdm.tqdm(df_outer.iterrows(), total=len(df_outer)):
    if pd.notna(row["Name_y"]):
        merged_text = ""
        if pd.notna(row["Name_y"]):
            merged_text += row["Name_y"] + '\n'
        if pd.notna(row["Content_y"]):
            merged_text += row["Content_y"] + '\n'
        if pd.notna(row["Weekly_y"]):
            merged_text += row["Weekly_y"]

        embedding = model.encode(merged_text)
        names.append(row["Name_y"])
    else:
        merged_text = ""
        if pd.notna(row["Name_x"]):
          merged_text += row["Name_x"] + '\n'
        if pd.notna(row["Content_x"]):
          merged_text += row["Content_x"] + '\n'
        if pd.notna(row["Weekly_x"]):
          merged_text += row["Weekly_x"]

        embedding = model.encode(merged_text)
        names.append(row["Name_x"])

    embeddings.append(embedding)
```

```python
from sklearn.decomposition import PCA
```

```python
from sklearn.decomposition import PCA
import numpy as np

embeddings_array = np.array(embeddings)

pca = PCA(n_components=2)
embeddings_2d = pca.fit_transform(embeddings_array)

embedding_df_2d = pd.DataFrame({'Name': names, 'Embedding_1': embeddings_2d[:, 0], 'Embedding_2': embeddings_2d[:, 1]})
```

```python
import plotly.express as px

# Assuming 'embedding_df_2d' contains the 2D embeddings and names
fig = px.scatter(embedding_df_2d, x='Embedding_1', y='Embedding_2', hover_data=['Name'], title='2D Embeddings Scatter Plot')

# Show the plot
fig.show()
```

```python
import plotly.express as px
from sklearn.manifold import TSNE

# Assuming 'embeddings' is your array of embeddings with shape (n_samples, 320)
embeddings_array = np.array(embeddings)

# Apply t-SNE to reduce to 2 dimensions
tsne = TSNE(n_components=2, random_state=42)
embeddings_2d = tsne.fit_transform(embeddings_array)

# Create a DataFrame for the 2D embeddings
embedding_df_2d = pd.DataFrame({'Name': names, 'Embedding_1': embeddings_2d[:, 0], 'Embedding_2': embeddings_2d[:, 1]})

# Plot 2D embeddings with hover labels
fig = px.scatter(embedding_df_2d, x='Embedding_1', y='Embedding_2', hover_data=['Name'], title='t-SNE Embeddings Scatter Plot')

# Show the plot
fig.show()
```
